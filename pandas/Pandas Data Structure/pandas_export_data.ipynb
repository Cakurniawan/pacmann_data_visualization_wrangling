## Mengeksport Data

Jika hasil olah data kita telah selesai dan data hasil olahan tersebut ingin disimpan dalam sebuah file, pandas juga menyediakan fungsi untuk menyimpan data tersebut dalam sebuah file. Kita bisa menggunakan fungsi `to_csv()` untuk mengeksport dataframe menjadi suatu file csv.

```python
import pandas as pd
```

```python
# buat data dalam dictionary
data = {
    "nama_barang": ["Indomie", "Taro", "Good Day", "Detergen"],
    "jenis_barang": ["food", "food", "food", "non-food"],
    "jumlah": [5, 2, 1, 1]
}

# buat dataframe dari data dictionary
df = pd.DataFrame(data)

# tampilkan data
df
```

Untuk mengeksport data dapat memanggil method `.to_csv()` dari dataframe.

```python
# eksport data ke file "data1.csv"
df.to_csv("data1.csv")
```

Hasil file dari script di bawah ini:

data1.jpg

Header dan index akan ikut diexport ke dalam file.

Jika tidak ingin mengeksport indexnya bisa menggunakan parameter `index=False`.

```python
# eksport data ke file "data1.csv" tanpa index
df.to_csv("data2.csv", index=False)
```

 Hasilnya akan seperti ini:

data2.jpg


Jika tidak ingin mengeksport headernya bisa menggunakan parameter `header=False`.

```python
# eksport data ke file "data1.csv" tanpa header
df.to_csv("data3.csv", header=False)
```

Hasilnya akan seperti ini:

data3.jpg
