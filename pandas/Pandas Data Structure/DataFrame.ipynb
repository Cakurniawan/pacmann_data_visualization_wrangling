### DataFrame

DataFrame merupakan sturuktur data pandas yang merepresentasikan bentuk tabel yang terdiri dari baris-baris dan kolom-kolom. 

dataframe.jpg


**Membuat DataFrame**

Untuk membuat tabel menggunakan DataFrame, bisa menggunakan fungsi `pd.DataFrame()`. Cara yang paling umum untuk membuat dataframe yaitu menggunakan data dictionary yang berisi list dengan panjang yang sama.
Misalnya data transaksi di bawah ini.

|kode_barang|nama_barang|jenis_barang|jumlah|
|:--|--|--|--:|
|IDMF|Indomie|Food|5|
|TROF|Taro|Food|2|
|GODF|Good Day|Food|1|
|DETN|Detergen|Non-food|1|

```python
import pandas as pd
```

```python
# buat data dalam bentuk data dictionary python
dict_data = {
    "nama_barang": ["Indomie", "Taro", "Good Day", "Detergen"],
    "jenis_barang": ["food", "food", "food", "non-food"],
    "jumlah": [5, 2, 1, 1]
}

# buat dataframe dari data dictionary
df = pd.DataFrame(dict_data)

# tampilkan data
df
```

Index baris akan otomatis menggunakan angka mulai dari 0. Tapi, kita juga dapat mengubahnya secara manual menggunakan parameter `index` pada fungsi `pd.DataFrame()`

```python
# buat data dalam bentuk data dictionary python
dict_data = {
    "nama_barang": ["Indomie", "Taro", "Good Day", "Detergen"],
    "jenis_barang": ["food", "food", "food", "non-food"],
    "jumlah": [5, 2, 1, 1]
}

# buat index data dalam bentuk python list
index_data = ["IDMF", "TROF", "GODF", "DETN"]

# buat dataframe dari data dictionary
df2 = pd.DataFrame(dict_data, index=index_data)

# tampilkan data
df2
```

DataFrame dan Series saling berhubungan. Jika kita ingin melihat hanya salah satu kolom dari dataframe, hasilnya akan ditampilkan dalam bentuk series. Jadi, dapat dikatakan bahwa DataFrame merupakan gabungan kolom dari satu atau lebih data series.

```python
# tampilkan kolom nama_barang dalam bentuk series
df2["nama_barang"]
```

```python
# tampilkan jenis data kolom nama_barang
type(df2["nama_barang"])
```

**Akses DataFrame**

Pada dataframe, terdapat **baris**, **kolom**, dan **cell**. Masing-masing punya cara mengakses yang berbeda-beda. Secara general kita bisa mengakses data dengan cara

* `dataframe.loc[[namaindex], [namakolom]]` 

*Akses Baris*

Mengakses baris pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[[namaindex]]` 

Pada fungsi `loc`, parameternya menggunakan nama index. Misalnya data Good Day itu nama indexnya "GODF". 

```python
# akses baris menggunakan fungsi loc
df2.loc[["GODF"]] 
```

```python
# akses lebih dari 1 baris menggunakan fungsi loc
df2.loc[["GODF","DETN"]] 
```

*Akses Kolom*

Mengakses kolom pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[:,[namakolom]]` 

Fungsi `loc` membutuhkan *nama kolom*.

Hal lain yang menjadi pembeda yaitu tanda `:`. Tanda itu berarti "select semua baris".

```python
# akses kolom menggunakan fungsi loc
df2.loc[:,["nama_barang"]] 
```

```python
# akses lebih dari 1 kolom menggunakan fungsi loc
df2.loc[:,["nama_barang","jumlah"]] 
```

*Akses Cell*

Mengakses cell pada dataframe dapat dilakukan  dengan cara:

* `dataframe.loc[[namaindex],[namakolom]]` 

Fungsi `loc` membutuhkan nama kolom/baris.

```python
# akses cell menggunakan fungsi loc
df2.loc[["GODF"],["nama_barang"]] 
```

```python
# akses lebih dari 1 cell menggunakan fungsi loc
df2.loc[["GODF","DETN"],["nama_barang","jumlah"]] 
```

**Update data pada DataFrame**

Untuk melakukan update data pada dataframe, kita harus tahu cara mengakses data. Setelah mengakses data, data baru dapat ditimpa ke data lama. 

*Update baris*

Untuk melakukan update pada baris tertentu, dapat menggunakan pola `baris_yang_ingin_diupdate = listdata`. Misalnya ingin mengupdate data pada baris "GODF", akses terlebih dahulu baris "GODF" kemudian beri list data updatenya. Perlu diperhatikan bahwa panjang listnya harus sesuai dengan banyaknya kolom.

```python
# tampilkan data
df2
```

```python
# update baris GODF menggunakan fungsi loc
df2.loc["GODF"] = ["GG Day", "Food", "3"] 

# tampilkan data
df2
```

*Update kolom*

Untuk melakukan update pada kolom tertentu, dapat menggunakan pola `kolom_yang_ingin_diupdate = listdata`. Misalnya ingin mengupdate data pada kolom "jenis_barang", akses terlebih dahulu kolom "jenis_barang" kemudian beri list data updatenya. Perlu diperhatikan bahwa panjang listnya harus sesuai dengan banyaknya baris.

```python
# update nilai kolom jumlah
df2.loc[:,"jumlah"] = [5, 3, 2, 2] 

# tampilkan data
df2
```

*Update cell*

Untuk melakukan update pada baris tertentu, dapat menggunakan pola `cell_yang_ingin_diupdate = data`. Misalnya ingin mengupdate data pada baris "IDMF" kolom "nama_barang", akses terlebih dahulu baris "IDMF" dan kolom "nama_barang" kemudian beri data updatenya. 

```python
# update nilai baris TROF kolom nama_barang menjadi Taro SW
df2.loc["TROF","nama_barang"] = "Taro SW" 

# tampilkan data
df2
```

```python
# update nilai multiple cell
df2.loc[["DETN","GODF"],["jenis_barang","jumlah"]] = [["non-food","1"],["food","3"]] 

# tampilkan data
df2
```

**Tambah data pada DataFrame**

Pada dataframe, kita dapat menambah baris atau kolom
* Menambah baris: `dataframe =  dataframe.append(newdata)`
* Menambah kolom: `dataframe[namakolom] =  newdata`

```python
# menambah baris
# buat baris dalam bentuk dataframe
new_row = pd.DataFrame({
    "nama_barang": "Pasta Gigi", 
    "jenis_barang": "non-food", 
    "jumlah": 4}, 
    index=["PGIN"]
    )

# tambahkan data baru ke dataframe yang telah dibuat sebelumnya
df2 = df2.append(new_row)

# tampilkan data
df2
```

```python
# menambah kolom
# tambahkan kolom baru dengan nama "harga_barang"
df2["harga_barang"] = [2500, 6000, 9000, 4500, 12000]

# tampilkan data
df2
```

**Menghapus data pada DataFrame**

Pada dataframe, kita dapat menghapus baris atau kolom menggunakan fungsi `drop()`.

* Menghapus baris: `dataframe = dataframe.drop(namaindex)`
* Menghapus kolom: `dataframe = dataframe.drop(namaindex, axis=1)`

```python
# menghapus baris
# menghapus baris pada index PGIN
df2 = df2.drop(["PGIN"])

# tampilkan data
df2
```

```python
# menghapus kolom
# menghapus kolom harga_barang
df2 = df2.drop(["harga_barang"], axis=1)

# tampilkan data
df2
```
